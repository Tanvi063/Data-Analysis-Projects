```python
df['year'] = pd.to_numeric(df['year'], errors='coerce', downcast='integer')

# Plotting the bar plot using value counts
plt.figure(figsize=(10, 6))
df['year'].value_counts().sort_index().plot(kind='line', color='r')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.title('Number of shows over the year', fontsize=25, fontweight="bold")
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Movie/TV shows', fontsize=15)
plt.xticks(rotation=90)

# Assuming "year" column is present in your DataFrame
sns.countplot(data=df[df['year'] > 2005], x='year', palette='viridis')

plt.show()
```

```python
plt.figure(figsize=(16, 8))
sns.set_style('darkgrid')
plt.title("Top 10 genres present in Hotstar", fontsize=25, fontweight='bold')
plt.xlabel('Genre', fontsize=15)
plt.xticks(rotation=75)
plt.ylabel('Count', fontsize=15)

# Assuming "genre" column is present in your DataFrame
data = df['genre'].value_counts().nlargest(10)
sns.barplot(x=data.index, y=data)

plt.show()
```
