### Popularity Based Recommendation System

```python
#top 50 books who has minimum rating of 250 votes now we will meerge te dataset of ratings and books
```

```python
ratings_with_bkname=ratings.merge(books,on='ISBN')
```

```python
ratings_with_bkname
```

```python
num_rating_df=ratings_with_bkname.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_bkname.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

### Collaborative Filtering Based Approach

```python
#filtering based on users
```

```python
x=ratings_with_bkname.groupby('User-ID').count()['Book-Rating']>200
who_read_books=x[x].index
```

```python
filtered_rating=ratings_with_bkname[ratings_with_bkname['User-ID'].isin(who_read_books)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index  #boolean indexing
```

```python
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt) #euclidean dist
```

```python
similarity_score.shape
```

```python
#function we wll give book name and based on that book name 5 books will be suggested
```

```python
np.where(pt.index=='1984')[0][0]
```

```python
index=np.where(pt.index=='Zoya')[0][0]
```

```python
list(enumerate(similarity_score[0]))    #book 1984 ka 1st book ke sath similarity score 0.99 hai...and rest
```

```python
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True)[1:6]  #not taken from 0 cuz 1st one is book itself
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Message in a Bottle')
```
